Customer_type = These describes the level of the customer either membership or Normal

Gender = sex of the customer either male or female

Product_line = Objects and materials that was sold out

Unit_price = Price of each products in $

Quantity = Numbers of products purchased by customers 

Tax = 5% tax rate from customers 

Total = Total price taxes included 

Date = Date which goods were purchased

Time = purchased time 

Payment = Different methods of payments (E wallet, cash or credit card)

cogs = Cost of goods sold 

gross margin percentage = gross margin percentage

```python
#quantity sold based on valuation rounded
Product_line = ['Electronic accessories', 'Health and beauty', 'Home and lifestyle', 
               'Sports and travel', 'Food and beverages', 'Fashion accessories']
df_rating = df.round()
for line in Product_line:
    df_line_3 = df_rating[df_rating ['Product_line']==line]
    df_amount_rating_Product_line = pd.DataFrame(df_line_3.groupby('Rating')['Quantity'].sum())
    df_amount_rating_Product_line = df_amount_rating_Product_line.reset_index()
    print(f'Quantity sold based on the valuation of the product_line:{line}')
    plt.figure(figsize=(6,4))
    sns.barplot(x= 'Rating', y='Quantity', data=  df_amount_rating_Product_line)
    plt.grid()
    plt.show()
```
